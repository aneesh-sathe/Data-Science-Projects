## Importing Libraries & Dataset

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("children.csv")
df2 = pd.read_csv("women.csv")
```

```python
df1.head()
```

```python
df2.head()
```

## Dropping Unwanted Series Objects 

```python
df1.drop(columns = ["S.No", "State/UT"], inplace = True)
df2.drop(columns = ["S.No", "State/UT"], inplace = True)
```

```python
df1.columns
```

## Replacing "-" in the data with 0

```python
df1[["2019","2020","2021"]] = df1[["2019","2020","2021"]].replace("-", 0)
df1
```

```python
df2[["2019","2020","2021"]] = df2[["2019","2020","2021"]].replace("-", 0)
df2
```

## Annotating Women's Data with (W)

```python
for x in df2.columns[1:]:
    y = x + "(W)"
    df2 = df2.rename(columns={x:y})
    
df2
```

## Converting Data Type to Numeric 

```python
for c in df1.columns[1:]:
    df1[c] = pd.to_numeric(df1[c])
```

```python
for c in df2.columns[1:]:
    df2[c] = pd.to_numeric(df2[c])
```

## Normalizing the Data (Min-Max Scaling)

```python
def min_max(x,xmax,xmin):
    x1 = (x - xmin)/(xmax-xmin)
    return x1 * 100

for c in df2.columns[1:]:
    xmax = df2[c].max()
    xmin = df2[c].min()
    df2[c] = df2[c].apply(lambda x : min_max(x,xmax,xmin) )
    
for c in df1.columns[1:]:
    xmax = df1[c].max()
    xmin = df1[c].min()
    df1[c] = df1[c].apply(lambda x : min_max(x,xmax,xmin) )
```

## Removing Special Symbols from State Names (*,+,@,..)


```python
def rep(x):
    if ("*" in x ):
        x1 = x.replace("*", "")
    elif ("@" in x):
        x1 = x.replace("@", "")
    elif ("+" in x):
        x1 = x.replace("+", "")
    else:
        x1 = x
    return x1
        

df1["Name"] = df1["Name"].apply(lambda x : rep(x))
df2["Name"] = df2["Name"].apply(lambda x : rep(x))
```

```python
df2
```

## Merging the Two Data Sets into One 

```python
df = pd.merge(df2,df1, on = "Name")
df
```

## Calculating the Average Crime from 2019-21

```python
df["Mean Crime"] = df[["2019", "2020", "2021"]].mean(axis=1)
df["Mean Crime(W)"] = df[["2019(W)", "2020(W)", "2021(W)"]].mean(axis=1)
df.head()
```

## Crime WRT Population (C/P):

```python
df["C/P"] = df["Mean Crime"]/df["midPopulation"]
df["C/P(W)"] = df["Mean Crime(W)"]/df["midPopulation(W)"]
df.head()
```

## Top 5 states with highest crimes against women

```python
cr_w = df.sort_values("Mean Crime(W)", ascending = 0)
cr_w["Name"].head()
```

## Top 5 states with Highest Crimes Against Children 

```python
cr_c = df.sort_values("Mean Crime", ascending = 0)
cr_c["Name"].head()
```

## Top 5 states with Highest Crimes Against Women WRT Population

```python
cr_wp = df.sort_values("C/P(W)", ascending = 0)
cr_wp["Name"].head()
```

## Top 5 states with Highest Crimes Against Children WRT Population

```python

cr_cp = df.sort_values("C/P", ascending = 0)
cr_cp["Name"].head()
```

## States with Highest Crime Against Children AND Women 

```python
cr_cw = df.loc[(df["Mean Crime"] > 50) & (df["Mean Crime(W)"] > 50)]
cr_cw.sort_values(["Mean Crime", "Mean Crime(W)"], ascending =(0,0))
cr_cw[["Name", "Mean Crime","Mean Crime(W)"]]
```

## Safest States for Both Women and Children

```python
safe = df.loc[(df["Mean Crime"] < 10) & (df["Mean Crime(W)"] < 10)]
safe.sort_values(["Mean Crime", "Mean Crime(W)"], ascending =(1,1))
safe[["Name", "Mean Crime","Mean Crime(W)"]].head()
```

## Crimes against Women vs Children Population 

```python
c_w = df["Mean Crime(W)"]
pop_c = df["midPopulation"]
state = df["Name"]
plt.figure(figsize=(10,5))
plt.xticks(rotation = 90)
plt.plot(state, c_w, "r", label = "Mean Crime (Women)")
plt.plot(state, pop_c, "b", label = "Child Population")
plt.title("Crimes Against Women vs Child Population (Statewise)")
plt.grid()
plt.legend()
plt.show()
```

## Crimes Against Women vs Crimes Against Children from 2019-21

```python
n = df["Name"]
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(15,5), sharex=True)
fig.suptitle("Crimes Against Women vs Children (2019-21)")

ax1.set_xticklabels(n,rotation=90)
ax1.plot(n, df["2019"], "r", label = "2019")
ax1.plot(n, df["2020"], "g", label = "2020")
ax1.plot(n, df["2021"], "b", label = "2021")
ax1.grid()

ax2.set_xticklabels(n,rotation=90)
ax2.plot(n, df["2019(W)"], "r", label = "2019(W)")
ax2.plot(n, df["2020(W)"], "g", label = "2020(W)")
ax2.plot(n, df["2021(W)"], "b", label = "2021(W)")
ax2.grid()

ax1.set_title("Children")
ax2.set_title("Women")
ax1.legend()
ax2.legend()
plt.show()
```
